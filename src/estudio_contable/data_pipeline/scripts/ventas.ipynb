```python
csv_path = Path(r'E:\Desktop\Archivos\Codigo\Python\automatizaciones_ec\src\estudio_contable\data_pipeline\data\raw\ventas\mis_comprobantes\CSV')
all_csvs = csv_path.glob('*.zip')

def extraer_del_zip(folder: Path):
    zip_files = list(folder.glob('*.zip'))
    if not zip_files:
        print("No hay un zip acá")
        return
    
    # zip_path = Zip File
    zip_path = zip_files[0]

    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(path=folder)
    
    zip_path.unlink()

def extraer_muchos_zips(folder: Path):
    for _ in folder.glob('*.zip'):
        extraer_del_zip(folder)
```

```python
archivo_venta_path_xlsx = 'path_al_xlsx_file_de_ventas'

def afip_pre_procesamiento_mis_comprobantes_ventas_xlsx(file_path):
    # Renombre de columnas
    columnas = {
        'Fecha'                 : 'fecha',
        'Tipo'                  : 'tipo_comprobante',
        'Punto de Venta'        : 'punto_de_venta',
        'Número Desde'          : 'nro_de_comprobante_desde',
        'Número Hasta'          : 'nro_de_comprobante_hasta',
        'Cód. Autorización'     : 'cod_autorizacion',
        'Tipo Doc. Receptor'    : 'tipo_doc_receptor',
        'Nro. Doc. Receptor'    : 'nro_doc_receptor',
        'Denominación Receptor' : 'receptor',
        'Tipo Cambio'           : 'tipo_de_camio',
        'Moneda'                : 'moneda',
        'Imp. Neto Gravado'     : 'neto_gravado',
        'Imp. Neto No Gravado'  : 'no_gravado',
        'Imp. Op. Exentas'      : 'exento',
        'Otros Tributos'        : 'otros_tributos',
        'IVA'                   : 'iva',
        'Imp. Total'            : 'importe_total'
    }

    # Columnas que hay que aclarar el type
    tipos_columnas = {
        'Punto de Venta'     : str,
        'Número Desde'       : str,
        'Número Hasta'       : str,
        'Cód. Autorización'  : str,
        'Nro. Doc. Receptor' : str,
    }

    df = pd.read_excel(file_path, dtype=tipos_columnas) # Lectura del archivo
    cuit_cliente = df.iloc[0].index[0][-11:]            # Agarramos el cuit del cliente usando la primera línea del excel
    df.columns = df.iloc[0]                             # Ponemos las columnas reales como columnas (antes estaban como primer registro)
    df = df[2:].reset_index(drop=True)                  # Arrancamos el df en el segundo registro (acá empiezan los registros en realidad)
    df = df.rename(columns=columnas)                    # Renombramos las columnas para nuestra base de datos
    df['cuit_cliente'] = cuit_cliente                   # Agregamos el cuit del cliente que obtuvimos antes para identificar de quién son las ventas

    return df

afip_pre_procesamiento_mis_comprobantes = {
    ('ventas', 'excel') : afip_pre_procesamiento_mis_comprobantes_ventas_xlsx
}

df = afip_pre_procesamiento_mis_comprobantes[('ventas', 'excel')](archivo_venta_path_xlsx)
df.head(5)
```
